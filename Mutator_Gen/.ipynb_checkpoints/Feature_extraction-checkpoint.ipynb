```python
import pandas as pd
import re
import uuid
import sqlite3
from langdetect import detect, LangDetectException

###############################################################
# CONFIG
###############################################################

INPUT_FILE = "mutation_1.xlsx"     # parent dataset
OUTPUT_DB = "genome_db.sqlite"
OUTPUT_CSV = "genome_db.csv"

MIN_TOKEN_LENGTH = 30
MAX_TOKEN_LENGTH = 200


###############################################################
# PROMPT NORMALIZATION
###############################################################

def normalize_prompt(prompt: str) -> str:
    prompt = prompt.lower()
    prompt = re.sub(r'\s+', ' ', prompt)
    prompt = prompt.strip()
    return prompt


###############################################################
# TECHNIQUE DETECTION
###############################################################

def detect_technique(prompt): #checks for any similar sentences and classifies based on that

    if "ignore previous instructions" in prompt:
        return "instruction_override"

    if "ignore all previous instructions" in prompt:
        return "instruction_override"

    if "imagine you are" in prompt or "pretend to be" in prompt:
        return "role_play"

    if "translate" in prompt:
        return "translation"

    if "decode" in prompt:
        return "decoding"

    if "encode" in prompt:
        return "encoding_attack"

    if "explain step by step" in prompt:
        return "chain_of_thought"

    if "json" in prompt:
        return "json_structured"

    return "generic_request"


###############################################################
# PERSONA DETECTION
###############################################################

def detect_persona(prompt):

    patterns = [
        r"you are a ([a-z ]+)", #uses regex aka (a+z)*
        r"act as a ([a-z ]+)",
        r"pretend to be a ([a-z ]+)"
    ]

    for pattern in patterns:
        match = re.search(pattern, prompt)
        if match:
            persona = match.group(1).strip()
            return persona.replace(" ", "_")

    return "none"


###############################################################
# FRAMING DETECTION
###############################################################

def detect_framing(prompt):

    if "hypothetical" in prompt or "imagine" in prompt:
        return "hypothetical"

    if "research" in prompt or "study" in prompt:
        return "research"

    if "debug" in prompt:
        return "debugging"

    if "fictional" in prompt or "story" in prompt:
        return "story"

    return "direct_command"


###############################################################
# ENCODING DETECTION
###############################################################

def detect_encoding(prompt): #this is only for those few prompts where they ask the LLM to decode some encoded message like "decode this base64 code"

    if "base64" in prompt:
        return "base64"

    if "hex" in prompt:
        return "hex"

    if "rot13" in prompt:
        return "rot13"

    if "encode" in prompt:
        return "encoding"

    return "none"


###############################################################
# STRUCTURE DETECTION
###############################################################

def detect_structure(prompt): #self explanatory

    if "{" in prompt and "}" in prompt:
        return "json"

    if "?" in prompt:
        return "question"

    if len(prompt.split(".")) > 1:
        return "multi_sentence"

    return "plain_text"


###############################################################
# NESTING DEPTH
###############################################################

def detect_nesting_depth(prompt):

    quote_count = prompt.count('"') + prompt.count("'")

    if quote_count >= 4:
        return 2

    if quote_count >= 2:
        return 1

    return 0


###############################################################
# LANGUAGE DETECTION
###############################################################

def detect_language(prompt):

    try:
        return detect(prompt)
    except LangDetectException:
        return "unknown"


###############################################################
# COMPLEXITY SCORE
###############################################################

def compute_complexity(genome):

    score = 0 #basically just incrementing the severity

    if genome["persona"] != "none":
        score += 1

    if genome["encoding"] != "none":
        score += 1

    if genome["nesting_depth"] > 0:
        score += 1

    if genome["technique"] == "role_play":
        score += 1

    if genome["structure"] == "multi_sentence":
        score += 1

    return score


###############################################################
# GENOME EXTRACTION
###############################################################

def extract_genome(prompt_text, prompt_id):

    normalized = normalize_prompt(prompt_text)

    tokens = len(normalized.split())

    if tokens < MIN_TOKEN_LENGTH or tokens > MAX_TOKEN_LENGTH:
        return None

    genome = {}

    genome["genome_id"] = str(uuid.uuid4())
    genome["prompt_id"] = prompt_id

    genome["technique"] = detect_technique(normalized)
    genome["persona"] = detect_persona(normalized)
    genome["framing"] = detect_framing(normalized)
    genome["encoding"] = detect_encoding(normalized)
    genome["language"] = detect_language(normalized)
    genome["nesting_depth"] = detect_nesting_depth(normalized)
    genome["structure"] = detect_structure(normalized)

    genome["complexity_score"] = compute_complexity(genome)

    return genome


###############################################################
# LOAD DATASET
###############################################################

def load_parent_dataset():

    df = pd.read_excel(INPUT_FILE)

    if "prompt_text" not in df.columns:
        raise ValueError("Dataset must contain 'prompt_text' column")

    return df


###############################################################
# GENOME DATABASE (SQLite)
###############################################################

def initialize_db():

    conn = sqlite3.connect(OUTPUT_DB)

    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS genomes (
        genome_id TEXT PRIMARY KEY,
        prompt_id TEXT,
        technique TEXT,
        persona TEXT,
        framing TEXT,
        encoding TEXT,
        language TEXT,
        nesting_depth INTEGER,
        structure TEXT,
        complexity_score INTEGER
    )
    """)

    conn.commit()

    return conn


###############################################################
# MAIN EXTRACTION PIPELINE
###############################################################

def run_extraction():

    df = load_parent_dataset()

    conn = initialize_db()
    cursor = conn.cursor()

    genomes = []

    for idx, row in df.iterrows():

        prompt = str(row["prompt_text"])
        prompt_id = row.get("prompt_id", idx)

        genome = extract_genome(prompt, prompt_id)

        if genome is None:
            continue

        genomes.append(genome)

        cursor.execute("""
        INSERT OR IGNORE INTO genomes VALUES (?,?,?,?,?,?,?,?,?,?)
        """, (
            genome["genome_id"],
            genome["prompt_id"],
            genome["technique"],
            genome["persona"],
            genome["framing"],
            genome["encoding"],
            genome["language"],
            genome["nesting_depth"],
            genome["structure"],
            genome["complexity_score"]
        ))

    conn.commit()

    genome_df = pd.DataFrame(genomes)

    genome_df.to_csv(OUTPUT_CSV, index=False)

    conn.close()

    print("Extraction complete.")
    print(f"Genomes created: {len(genomes)}")
    print(f"Saved to {OUTPUT_DB} and {OUTPUT_CSV}")


###############################################################
# RUN
###############################################################

if __name__ == "__main__":
    run_extraction()
```
